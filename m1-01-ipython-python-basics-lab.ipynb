```python
events = [
    {"user_id": 1, "event_type": "click", "duration_seconds": 30},
    {"user_id": 2, "event_type": "view", "duration_seconds": 45},
    {"user_id": 3, "event_type": "click", "duration_seconds": -5},
    {"user_id": 1, "event_type": "scroll", "duration_seconds": 10},
    {"user_id": 4, "event_type": "view", "duration_seconds": "bad"},
    {"user_id": 2, "event_type": "click", "duration_seconds": 60},
    {"user_id": 5, "event_type": "scroll", "duration_seconds": 20},
    {"user_id": 3, "event_type": "view", "duration_seconds": 15},
    {"user_id": 6, "event_type": "click", "duration_seconds": 90},
    {"user_id": 7, "event_type": "view", "duration_seconds": 25},
    {"user_id": 8, "event_type": "scroll", "duration_seconds": 40},
    {"user_id": 9, "event_type": "click", "duration_seconds": 50},
]
cleaned_events = []
for event in events:
    d = event["duration_seconds"]
    if isinstance(d,(int,float)) and d > 0:
        new_event = event.copy()
        new_event ["duration_minutes"] = d / 60
        cleaned_events.append(new_event)
print(events)
print(cleaned_events)
#"events"  is generated by AI.
```

```python
events = [
    {"event_type": "login", "duration_minutes": 2, "user_id": 1},
    {"event_type": "login", "duration_minutes": 3, "user_id": 2},
    {"event_type": "purchase", "duration_minutes": 10, "user_id": 1},
    {"event_type": "logout", "duration_minutes": 1, "user_id": 1},
    {"event_type": "purchase", "duration_minutes": 8, "user_id": 3},
]
event_counts={}
for event in events:
    event_type = event["event_type"]
    event_counts[event_type] = event_counts.get(event_type,0) + 1
print(event_counts)
```

```python
duration_totals = {}
duration_counts = {}

for event in events:
    event_type = event["event_type"]
    duration_totals[event_type] = duration_totals.get(event_type, 0) + event["duration_minutes"]
    duration_counts[event_type] = duration_counts.get(event_type, 0) + 1

avg_durations = {}

for event_type in duration_totals:
    avg_durations[event_type] = duration_totals[event_type] / duration_counts[event_type]
print(avg_durations)
```

```python
unique_users = set()

for event in events:
    unique_users.add(event["user_id"])

unique_user_count = len(unique_users)
print(unique_user_count)
```

```python
total_events = 0
for c in event_counts.values():
    total_events += c
if total_events == len(events):
    print("True")
else:
    print("False")

```
